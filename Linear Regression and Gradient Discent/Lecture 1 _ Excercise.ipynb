```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
# from sklearn.linear_model import LinearRegression
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Data Sets Lecture 1 Exercise Code Basics.csv")
data.head()
```

```python
plt.scatter(data['year'],data['per_capita_income_(US$)'],marker='+',c ='b',label="income")
plt.title(" Per Capita Income (Year)",c='r')
plt.xlabel(" Years ",c='r')
plt.ylabel(" Per Capita Income in USD",c='r')
plt.legend()
plt.show()
```

```python
year = data[['year']]
income = data['per_capita_income_(US$)']
```

```python
model= linear_model.LinearRegression()
model.fit(year,income)
```

```python
model.predict([[2022]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
828.465075*2022+(-1632210.7578554575)
```

```python
plt.scatter(data['year'],data['per_capita_income_(US$)'],marker='+',c ='b',label="income")
plt.plot(data['year'], model.predict(data[['year']]),label=" Predicted income ",c='r')
plt.title(" Per Capita Income (Year)",c='r')
plt.xlabel(" Years ",c='r')
plt.ylabel(" Per Capita Income in USD",c='r')
plt.legend()
plt.show()
```
