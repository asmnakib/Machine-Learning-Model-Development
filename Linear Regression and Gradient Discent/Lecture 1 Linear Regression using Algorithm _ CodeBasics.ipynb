```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
# from sklearn.linear_model import LinearRegression
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("House_price_prediction_codebasics_data_sets.csv")
data
```

```python
plt.scatter(data['Area'],data['Price'],marker="x", c="r",label ="Price Point")
# plt.plot(data['Area'],data['Price'],label ="Line", c='b')
plt.title("House Price Prediction",c = "y")
plt.ylabel(" Price (in USD)", c = 'r')
plt.xlabel("Size in SQFT",c='r')
plt.legend()

plt.show()
```

from the plot I can easily understand the type of data , it is suitable for linear regression

```python
# creating an object
model = linear_model.LinearRegression()
```

```python
model.fit(data[['Area']],data.Price)
```

```python
model.predict([[3000]])
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
plt.scatter(data['Area'],data['Price'],marker="x", c="r",label ="Price Point")
plt.plot(data['Area'],model.predict(data[['Area']]),label ="Predicted Linear Line", c='b')
plt.title("House Price Prediction",c = "y")
plt.ylabel(" Price (in USD)", c = 'r')
plt.xlabel("Size in SQFT",c='r')
plt.legend()

plt.show()
```
