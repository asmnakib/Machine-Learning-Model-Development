```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
# from sklearn.linear_model import LinearRegression
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df  = pd.read_csv("Data Sets Lecture 6 CODEBASICS.csv")
df
```

```python
plt.scatter(df['Mileage'],df['Sell Price($)'],marker ='+',c ='r')
plt.title("Mileage vs Sell Price",c ='b')
plt.show()
```

```python
plt.scatter(df['Age(yrs)'],df['Sell Price($)'],marker ='*',c ='r')
plt.title("Age  vs Sell Price",c ='b')
plt.show()
```

```python
x = df[['Mileage','Age(yrs)']]
y = df['Sell Price($)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
model =linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```
