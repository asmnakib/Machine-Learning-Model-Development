```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Data Sets Lecture 2 Code Basics.csv')
df.head()
```

```python
# this NaN should handeled before going further
mediaan= df.bedrooms.median()
mediaan
```

```python
# to make it integer
import math
median_bedrooms = math.floor(mediaan)
median_bedrooms
```

```python
# filling the median_bedrooms value by using fillna
df['bedrooms']=df['bedrooms'].fillna(median_bedrooms)
df
```

```python
plt.plot(df['area'],df['price'],c='r',label ='area vs price')
# plt.plot(df['bedrooms'],df['price'],c='b',label ='bedrooms vs price')
# plt.plot(df['age'],df['price'],c='y',label ='age vs price')
plt.legend()
plt.show()
```

```python
plt.plot(df['bedrooms'],df['price'],c='b',label ='bedrooms vs price')
plt.legend()
plt.show()
```

```python
plt.plot(df['age'],df['price'],c='y',label ='age vs price')
plt.legend()
plt.show()
```

```python
# Creating Model Object
model = linear_model.LinearRegression()
```

```python
# Creating iNdependent Variables
x = df[['area','bedrooms','age']]

# creating dependent Variable
y= df.price
```

```python
model.fit(x,y)
```

```python
b =model.intercept_
b
```

```python
coef = model.coef_
coef
```

```python
price = model.predict([[3000,3,40]])
print(price)
```
